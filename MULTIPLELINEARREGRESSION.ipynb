```python
# Supervised Machine Learning
# Regression Multiple Linear Regression
# Third Project : Predicting Profit Based On Multiple Independent Variable
# import all necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/Startups_Data.csv')
data
```

```python
x = ['R&D Spend'	,'Administration'	,'Marketing Spend',	'State']
x = data[x].values
x
```

```python
y = data.Profit.values
y
```

```python
# ONEHOTENCODING

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

transformer = ColumnTransformer(
    transformers = [('OneHot', OneHotEncoder(),[3])],   # List of (name, transformer, columns)
    remainder = 'passthrough'    # please do not remove non specified columns
)

x = transformer.fit_transform(x)
x = x.astype('int')
```

```python
x

# california    florida   newyork   r_d     admin   market
#   0,            0,          1,    165349, 136897, 471784,
#   1,            0,          0,    162597, 151377, 443898,

# newyork --> 0 0 1
# califronia --> 1 0 0
# florida --> 0 1 0
```

```python
# train test split
# 0.2 means 20% of the data I will be using for Testing
# remaining 80% of the data I will be using for Training

from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print('Original Shape')
print(x.shape)
print(y.shape)
print('After Splitting')
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
import sklearn.linear_model as lm
mlr = lm.LinearRegression()

mlr.fit(x_train,y_train)
```

```python
pred_values = mlr.predict(x_test)
print(pred_values)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,pred_values)
print(f'Regression Score : {score}')
```
